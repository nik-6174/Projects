```python
import csv
import random
import string

# Your existing code for Node and Graph classes here...

def generate_dummy_data():
    dummy_data = []
    # Generate dummy coordinates and names
    for i in range(1,101):  # Generate 10 nodes
        x = round(random.uniform(0, 100), 2)  # Random coordinates rounded to .2f precision
        y = round(random.uniform(0, 100), 2)
        name = ''.join(random.choices(string.ascii_uppercase + string.digits, k=5))  # Random name
        dummy_data.append((x, y, name))
    return dummy_data

def write_data_to_csv(filename, data):
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['X', 'Y', 'Name'])  # Write header
        writer.writerows(data)

if __name__ == "__main__":
    # Generate dummy data
    dummy_data = generate_dummy_data()
    # Write dummy data to CSV file
    write_data_to_csv('node_data.csv', dummy_data)
```
