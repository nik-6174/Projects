```python
import pandas as pd
import os
from geopy.distance import geodesic
import ast  # Library for handling string conversion to dictionary

# Define the file paths for nodes.csv and charging.csv within the 'data' folder
nodes_file = os.path.join('data', 'nodes.csv')
charging_file = os.path.join('data', 'charging.csv')

# Load nodes.csv and charging.csv into pandas DataFrames
nodes_df = pd.read_csv(nodes_file)
charging_df = pd.read_csv(charging_file)

# Function to extract coordinates from a string formatted as 'POINT (lon lat)'
def extract_coordinates(point_str):
    lon, lat = map(float, point_str.replace('POINT (', '').replace(')', '').split())
    return lat, lon

# Function to extract node coordinates from the 'data' column in nodes_df
def extract_node_coordinates(row):
    data_str = row['data']
    data_dict = ast.literal_eval(data_str)  # Convert string to dictionary
    return data_dict['y'], data_dict['x']  # Return latitude and longitude

# Function to find the nearest node ID for each charging station
def find_nearest_node(row):
    charging_lat, charging_lon = extract_coordinates(row['geometry'])  # Extract charging station coordinates
    min_distance = float('inf')
    nearest_node_id = None
    
    # Iterate through nodes to find the nearest node
    for index, node_row in nodes_df.iterrows():
        node_lat, node_lon = extract_node_coordinates(node_row)
        distance = geodesic((charging_lat, charging_lon), (node_lat, node_lon)).meters  # Calculate distance in meters
        
        if distance < min_distance:
            min_distance = distance
            nearest_node_id = node_row['node_id']
        if min_distance < 0.000001:
            break
    
    return nearest_node_id

# Apply the find_nearest_node function to each row in charging_df
charging_df['node_id'] = charging_df.apply(find_nearest_node, axis=1)

# Save the updated charging.csv DataFrame with the 'node_id' column
charging_df.to_csv(os.path.join('data', 'charging_with_nodes.csv'), index=False)

print('Node IDs assigned to charging stations successfully.') 
```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(coordinates1, coordinates2):
    # coordinates -> [latitude, longitude]
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [coordinates1[0], coordinates1[1], coordinates2[0], coordinates2[1]])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    # Radius of the Earth in kilometers (approximate)
    R = 6371.0

    # Calculate the distance
    distance = R * c
    return distance

# Example usage
latitude1, longitude1 = 28.6139, 77.2090  # Latitude and longitude of point 1 (e.g., Delhi)
latitude2, longitude2 = 19.0760, 72.8777  # Latitude and longitude of point 2 (e.g., Mumbai)

# Calculate the distance using the haversine_distance function
distance_km = haversine_distance([latitude1, longitude1], [latitude2, longitude2])
print(f"The distance between the two points is approximately {distance_km:.2f} kilometers.")
```

```python
# Define the Graph class
class Graph:
    def __init__(self):
        self.nodes = set()
        self.edges = defaultdict(dict)

    def add_node(self, node_id, longitude, latitude, street_count, highway):
        node = Node(node_id, longitude, latitude, street_count, highway)
        self.nodes.add(node)
        return node

    def add_edge(self, node_id1, node_id2, distance):
        node1.neighbors.add(node2)
        node2.neighbors.add(node1)
        self.edges[node_id1][node_id2] = distance
        self.edges[node_id2][node_id1] = distance
    
    def __getitem__(self, coordinates):
        return self.get_node[coordinates]
    
    def add_edge(self, node1, node2, distance):

        self.distance[node1][node2] = distance
        self.distance[node2][node1] = distance

    def nearest_node(self, coordinates):
        index = bisect.bisect_left(self.coordinates_list, coordinates)
        left_distance = abs(self.coordinates_list[index - 1] - coordinates)
        right_distance = abs(self.coordinates_list[index] - coordinates)
        if left_distance < right_distance:
            return self.coordinates_list[index - 1]
        else:
            return self.coordinates_list[index]

    def add_charging_nodes(self, charging_df):
        charging_nodes = set(charging_df['node_id'])
        for node in self.nodes:
            if node.id in charging_nodes:
                node.charge = True

    def haversine_distance(self, coordinates1, coordinates2):
        # Calculate the distance between two coordinates using the Haversine formula
        lat1, lon1 = coordinates1
        lat2, lon2 = coordinates2
        R = 6371.0  # Radius of the Earth in kilometers
        dlat = radians(lat2 - lat1)
        dlon = radians(lon2 - lon1)
        a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))
        distance = R * c
        return distance

    def build_graph(self, nodes_df, edges_df, charging_df):
        # Add nodes to the graph
        for idx, row in nodes_df.iterrows():
            node_id = row['node_id']
            longitude = row['longitude']
            latitude = row['latitude']
            street_count = row['street_count']
            highway = row['highway']
            node = self.add_node(node_id, longitude, latitude, street_count, highway)

        # Add charging information to nodes
        self.add_charging_nodes(charging_df)

        # Add edges to the graph
        for idx, row in edges_df.iterrows():
            node_id1 = row['node1']
            node_id2 = row['node2']
            coordinates1 = nodes_df[nodes_df['node_id'] == node_id1][['latitude', 'longitude']].values[0]
            coordinates2 = nodes_df[nodes_df['node_id'] == node_id2][['latitude', 'longitude']].values[0]
            distance = self.haversine_distance(coordinates1, coordinates2)
            self.add_edge(node_id1, node_id2, distance)

def shortest_path(self, source, destination):
    # Get the start and end nodes
    start_coordinates, end_coordinates = self.nearest_node(source), self.nearest_node(destination)
    start_node, end_node = self.get_node_from_coordinates[start_coordinates], self.get_node_from_coordinates[end_coordinates]

    # Function to calculate actual geographical distance between a node and the end node
    distance_from_end = lambda coordinates: self.haversine_distance(coordinates, end_coordinates)

    # Coefficients for the heuristic function
    h1 = 1  # Coefficient for the heuristic distance from the end point
    h2 = 1  # Coefficient for the distance
    full_charge = 350  # Charge in terms of distance (unit should be consistent with haversine_distance)

    # Find the shortest path using A* algorithm
    visited = {}
    queue = [[0, 0, full_charge, start_node, [start_node.id]]]  # [Heuristic value, Distance, Charge, Node, Path]
    while queue:
        heuristic, distance, charge, node, path = heapq.heappop(queue)
        if node == end_node:
            return distance, charge, path, len(queue)  # Return path information when destination is reached
        if node.charge:  # Fully charge at charging node (if applicable)
            charge = full_charge
            path[-1] = path[-1] + ' (charge)'
        if node not in visited or visited[node] > heuristic:
            visited[node] = heuristic
            for neighbor in node.neighbors:
                new_charge = charge - self.distance[node][neighbor]
                if new_charge < 10:  # Example threshold for minimal charge to proceed
                    continue
                neighbor_heuristic = h1 * (distance + self.distance[node][neighbor]) + h2 * distance_from_end(
                    neighbor.coordinates)
                heapq.heappush(queue, [neighbor_heuristic, distance + self.distance[node][neighbor], new_charge, neighbor, path + [neighbor.id]])
    return float('inf'), 0, [], 0  # Return if no path is found
```

```python
def astar_path(G, source, target, charging_distances, weight='length', max_range=15, charge=100):
    """Returns a list of nodes in a shortest path between source and target using the modified A* algorithm.

    There may be more than one shortest path.  This returns only one.

    Parameters
    ----------
    G : NetworkX graph

    source : node
        Starting node for path

    target : node
        Ending node for path
    
    charging_destinations : dict
        A dictionary containing the nodes that are closest to the charging stations
        & their distances from from the charging stations
    
    weight : string
        The edge data key corresponding to the edge weight
    
    max_range : int
        The maximum distance that the vehicle can travel on a single charge in kms
    
    charge : int
        The initial charge of the vehicle in percentage

    Raises NetworkXNoPath
        If no path exists between source and target.
    """

    if source not in G or target not in G:
        msg = f"Either source {source} or target {target} is not in G"
        raise nx.NodeNotFound(msg)

    # define function to find the distance from target node
    @lru_cache(maxsize=None)
    def distance_from_target(u, v):
        return geodesic((G.nodes[u]['x'], G.nodes[u]['y']), (G.nodes[v]['x'], G.nodes[v]['y'])).meters
    
    push = heappush
    pop = heappop
    weight = _weight_function(G, weight)

    G_succ = G._adj # For speed-up (and works for both directed and undirected graphs)

    # The queue stores priority, node, cost to reach, and parent.
    # Uses Python heapq to keep in priority order.
    # Add a counter to the queue to prevent the underlying heap from
    # attempting to compare the nodes themselves. The hash breaks ties in the
    # priority and is guaranteed unique for all nodes in the graph.

    def depth_first_search(source, target, max_range, charge=100):
        c = count()
        queue = [(0, charge, next(c), source, 0, None)]  # (heuristic val, charge, count, node, distance, parent)
        explored = {}
        enqueued = {}

        while queue:
            _, curcharge, _, curnode, dist, parent = pop(queue)

            # if we've reached the target
            if curnode == target:
                path = [curnode]
                node = parent
                while node is not None:
                    path.append(node)
                    node = explored[node]
                path.reverse()
                return path, dist

            # if the node is a charging node, fully charge the vehicle
            if curnode in charging_distances:
                # we recursively search for a path from charging node to the target node
                path, distance = depth_first_search(curnode, target, max_range, 100)
                if path and distance != float('inf'):
                    # add the path to the charging node
                    travelled_path = [curnode]
                    node = parent
                    while node is not None:
                        travelled_path.append(node)
                        node = explored[node]
                    travelled_path.reverse()
                    return travelled_path + path, dist + distance
            
            if curnode in explored:
                # Do not override the parent of starting node
                if explored[curnode] is None:
                    continue

                if enqueued[curnode] < curcharge:
                    continue
            
            explored[curnode] = parent

            for neighbor, w in G_succ[curnode].items():
                edge_weight = weight(curnode, neighbor, w)
                ncost = dist + edge_weight
                new_charge = curcharge - edge_weight/(max_range*10) # old charge - percentage battery used to travel edge_weight

                if new_charge < 0:
                    continue

                if neighbor in enqueued:
                    if enqueued[neighbor] <= ncost:
                        continue
                # find the distance from target node
                h = distance_from_target(neighbor, target)
                
                enqueued[neighbor] = ncost
                heuristic_val = ncost + h
                push(queue, (heuristic_val, new_charge, next(c), neighbor, ncost, curnode))
        return [], float('inf')
    
    return depth_first_search(source, target, max_range, charge)
```
